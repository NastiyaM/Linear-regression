```python
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
```

# Datasets

```python
# generate dataset
X, y = datasets.make_regression(n_samples=1_000, n_features=1, noise=10,shuffle=True)
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(2, 1, 1)
plt.title("Линейная модель") # заголовок
plt.xlabel("x") # ось абсцисс
plt.ylabel("y") # ось ординат
plt.grid()
plt.scatter(X, y, color='red', s=10)
plt.show()
```

# Linear Regression Scikit-learn

# Training and predictions

```python
# test and train data sizes
train_size = 800
test_size = 200

# split the data into training/testing sets
X_train = X[:train_size]
X_test = X[train_size:]

# split the targets into training/testing sets
y_train = y[:train_size]
y_test = y[train_size:]

# create linear regression object
regr = linear_model.LinearRegression()

# train the model using the training sets
regr.fit(X_train, y_train)

# make predictions using the testing set
y_pred = regr.predict(X_test)
```

# Graph Plotting 

```python
# plot outputs
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)
plt.title("Линейная модель") # заголовок
plt.xlabel("x") # ось абсцисс
plt.ylabel("y") # ось ординат
plt.grid()
plt.scatter(X_test, y_test, color='red', s=10)
plt.plot(X_test, y_pred, color='blue', linewidth=1)
plt.show()
```

# Metrics

```python
rmse = mean_squared_error(y_test, y_pred)
print('rmse',rmse)
r2 = r2_score(y_test, y_pred)
print('r2',r2)
```

# Linear Model

```python
from matplotlib import pyplot as plt
from celluloid import Camera
```

```python
class LinearRegressionUsingGD:
    """Linear Regression Using Gradient Descent.
    Parameters
    ----------
    eta : float
        Learning rate
    n_iterations : int
        No of passes over the training set
    Attributes
    ----------
    w_ : weights/ after fitting the model
    cost_ : total error of the model after each iteration
    """

    def __init__(self, eta=0.05, n_iterations=1000,n = 50):
        self.eta = eta
        self.n_iterations = n_iterations
        self.n = n
    def predict(self, x):
        """ Predicts the value after the model has been trained.
        Parameters
        ----------
        x : array-like, shape = [n_samples, n_features]
            Test samples
        Returns
        -------
        Predicted value
        """
        return np.dot(x, self.w_)

    def fit(self, x, y,X_test,X_test_,y_test):
        """Fit the training data
        Parameters
        ----------
        x : array-like, shape = [n_samples, n_features]
            Training samples
        y : array-like, shape = [n_samples, n_target_values]
            Target values
        Returns
        -------
        self : object
        """

        self.cost_ = []
        self.w_ = np.zeros((x.shape[1], 1))
        m = x.shape[0]
        fig = plt.figure()
        camera = Camera(fig)
        
        for _ in range(self.n):
            y_pred = np.dot(x, self.w_)
            residuals = y_pred - y
            gradient_vector = np.dot(x.T, residuals)
            self.w_ -= (self.eta / m) * gradient_vector
            cost = np.sum((residuals ** 2)) / (2 * m)
            self.cost_.append(cost)
            y_pred =self.predict(X_test)
            plt.title("Линейная модель") # заголовок
            plt.xlabel("x") # ось абсцисс
            plt.ylabel("y") # ось ординат
            plt.grid()
            plt.scatter(X_test_, y_test, color='red', s=10)
            t = plt.plot(X_test_, y_pred, color='blue', linewidth=1)
            plt.legend(t, [f'iteration{_}'])
            camera.snap()
        animation = camera.animate()
        animation.save('celluloid_legends.gif', writer='imagemagick')
        return self
    def fit_(self, x, y):
        """Fit the training data
        Parameters
        ----------
        x : array-like, shape = [n_samples, n_features]
            Training samples
        y : array-like, shape = [n_samples, n_target_values]
            Target values
        Returns
        -------
        self : object
        """

        self.cost_ = []
        self.w_ = np.zeros((x.shape[1], 1))
        m = x.shape[0]
        for _ in range(self.n_iterations):
            y_pred = np.dot(x, self.w_)
            residuals = y_pred - y
            gradient_vector = np.dot(x.T, residuals)
            self.w_ -= (self.eta / m) * gradient_vector
            cost = np.sum((residuals ** 2)) / (2 * m)
            self.cost_.append(cost)
        return self
```

```python
train_size = 800
test_size = 200
X_train = X[:train_size]
X_test_ = X[train_size:]
X_train = np.c_[np.ones((train_size, 1)),X_train]
X_test = np.c_[np.ones((test_size, 1)),X_test_]
# split the targets into training/testing sets
y_train = y[:train_size].reshape(train_size,1)
y_test = y[train_size:]
linear_regression_model = LinearRegressionUsingGD()
```
